```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data
```

## Player Count

* Display the total number of players


```python
user_count = len(purchase_data['SN'].value_counts())
user_count
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
unique_items = len(purchase_data['Item Name'].value_counts())
ave_price = round(purchase_data['Price'].mean(),2)
purchase_count = purchase_data['Item Name'].count()
total_rev = round(purchase_data['Price'].sum(),2)

summary_sale_df = pd.DataFrame ({
    'Number of Unique Items':  [unique_items],
    'Average Price ($)': [ave_price],
    'Number of Purchases': [purchase_count],
    'Total Revenue ($)': [total_rev],
        })
summary_sale_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
unique_users = purchase_data.drop_duplicates(subset=['SN'], keep='first', inplace=False)

gender_series = unique_users['Gender'].value_counts()

gender_dems = pd.DataFrame(data=gender_series)
gender_dems['Percentage'] = gender_dems['Gender']/user_count *100
gender_dems = gender_dems.rename(columns={'Gender': 'Count'})
gender_dems
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
gender_group = purchase_data.groupby('Gender')

purchase_count = gender_group.size()
purchase_count = purchase_count.rename('Purchase Count')
purchase_count_df = pd.DataFrame(data=purchase_count)

unique_gender = unique_users.groupby('Gender')
unique_count = unique_gender.size()
unique_count = unique_count.rename('Unique Users Count')
unique_count_df = pd.DataFrame(data=unique_count)

total_price = gender_group.sum()['Price']
total_price = total_price.rename('Total Price')
total_price_df = pd.DataFrame(data=total_price)

average_price = gender_group.mean()['Price']
average_price = average_price.rename('Average Price')
average_price_df = pd.DataFrame(data=average_price)

gender_price = pd.DataFrame.join(total_price_df, average_price_df, how='outer')
gender_price = pd.DataFrame.join(gender_price, purchase_count_df, how='outer')
gender_price = pd.DataFrame.join(gender_price, unique_count_df, how='outer')
gender_price["Average Total Purchase per Person"] = gender_price["Total Price"]/gender_price["Unique Users Count"]

gender_price
```

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
bins = (0,10, 14.5,  19.5, 24.5, 29.5, 34.5,39.5,200)
groups = purchase_data.groupby(pd.cut(purchase_data['Age'], bins=bins))

size = groups.size()
percent = size/size.sum()*100

size = size.rename('Total Count')
percent = percent.rename('Percentage of Players')

size_df = pd.DataFrame(data=size)
percent_df = pd.DataFrame(data=percent)

new_tab = pd.DataFrame.join(size_df, percent_df, how='outer' )
new_tab
```
